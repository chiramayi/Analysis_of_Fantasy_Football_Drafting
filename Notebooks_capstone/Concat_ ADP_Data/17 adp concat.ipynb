```python
import requests
from bs4 import BeautifulSoup
import pandas as pd
from urllib.request import urlopen as uReq


import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
import datetime as dt
from matplotlib import rcParams

%matplotlib inline
```

```python
adp_2017 = pd.read_csv('adp_2017.csv')
```

```python
wr_adp_17 = adp_2017.loc[adp_2017['Position'] == 'WR']
te_adp_17 = adp_2017.loc[adp_2017['Position'] == 'TE']
qb_adp_17 = adp_2017.loc[adp_2017['Position'] == 'QB']
pk_adp_17 = adp_2017.loc[adp_2017['Position'] == 'PK']
rb_adp_17 = adp_2017.loc[adp_2017['Position'] == 'RB']
def_adp_17 = adp_2017.loc[adp_2017['Position'] == 'Def']
```

```python
rb_adp_17['PositionRank'] = rb_adp_17['Rank'].rank(ascending = True)
rb_adp_17.sort_values(by = 'PositionRank')
wr_adp_17['PositionRank'] = wr_adp_17['Rank'].rank(ascending = True)
wr_adp_17.sort_values(by = 'PositionRank')
te_adp_17['PositionRank'] = te_adp_17['Rank'].rank(ascending = True)
te_adp_17.sort_values(by = 'PositionRank')
qb_adp_17['PositionRank'] = qb_adp_17['Rank'].rank(ascending = True)
qb_adp_17.sort_values(by = 'PositionRank')
def_adp_17['PositionRank'] = def_adp_17['Rank'].rank(ascending = True)
def_adp_17.sort_values(by = 'PositionRank')
pk_adp_17['PositionRank'] = pk_adp_17['Rank'].rank(ascending = True)
pk_adp_17.sort_values(by = 'PositionRank')
```

```python
adp_17_ranked = pd.concat([wr_adp_17, rb_adp_17, qb_adp_17, pk_adp_17, te_adp_17, def_adp_17])
adp_17_ranked = adp_17_ranked.sort_values(by = 'Rank')
```

```python
rb_adp_17.to_csv('adp_rb_2017.csv', index = False)
wr_adp_17.to_csv('adp_wr_2017.csv', index = False)
te_adp_17.to_csv('adp_te_2017.csv', index = False)
qb_adp_17.to_csv('adp_qb_2017.csv', index = False)
def_adp_17.to_csv('adp_def_2017.csv', index = False)
pk_adp_17.to_csv('adp_pk_2017.csv', index = False)

adp_17_ranked.to_csv('adp_ranked_2017.csv', index = False)
```
