```python
import requests
from bs4 import BeautifulSoup
import pandas as pd
from urllib.request import urlopen as uReq


import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
import datetime as dt
from matplotlib import rcParams

%matplotlib inline
```

```python
adp_2018 = pd.read_csv('adp_2018.csv')
```

```python
wr_adp_18 = adp_2018.loc[adp_2018['Position'] == 'WR']
te_adp_18 = adp_2018.loc[adp_2018['Position'] == 'TE']
qb_adp_18 = adp_2018.loc[adp_2018['Position'] == 'QB']
pk_adp_18 = adp_2018.loc[adp_2018['Position'] == 'PK']
rb_adp_18 = adp_2018.loc[adp_2018['Position'] == 'RB']
def_adp_18 = adp_2018.loc[adp_2018['Position'] == 'Def']
```

```python
rb_adp_18['PositionRank'] = rb_adp_18['Rank'].rank(ascending = True)
rb_adp_18.sort_values(by = 'PositionRank')
wr_adp_18['PositionRank'] = wr_adp_18['Rank'].rank(ascending = True)
wr_adp_18.sort_values(by = 'PositionRank')
te_adp_18['PositionRank'] = te_adp_18['Rank'].rank(ascending = True)
te_adp_18.sort_values(by = 'PositionRank')
qb_adp_18['PositionRank'] = qb_adp_18['Rank'].rank(ascending = True)
qb_adp_18.sort_values(by = 'PositionRank')
def_adp_18['PositionRank'] = def_adp_18['Rank'].rank(ascending = True)
def_adp_18.sort_values(by = 'PositionRank')
pk_adp_18['PositionRank'] = pk_adp_18['Rank'].rank(ascending = True)
pk_adp_18.sort_values(by = 'PositionRank')




```

```python
adp_18_ranked = pd.concat([wr_adp_18, rb_adp_18, qb_adp_18, pk_adp_18, te_adp_18, def_adp_18])
adp_18_ranked = adp_18_ranked.sort_values(by = 'Rank')
```

```python
rb_adp_18.to_csv('adp_rb_2018.csv', index = False)
wr_adp_18.to_csv('adp_wr_2018.csv', index = False)
te_adp_18.to_csv('adp_te_2018.csv', index = False)
qb_adp_18.to_csv('adp_qb_2018.csv', index = False)
def_adp_18.to_csv('adp_def_2018.csv', index = False)
pk_adp_18.to_csv('adp_pk_2018.csv', index = False)
```

```python
adp_18_ranked.to_csv('adp_ranked_2018.csv', index = False)
```
