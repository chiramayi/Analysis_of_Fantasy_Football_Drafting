```python
import requests
from bs4 import BeautifulSoup
import pandas as pd
from urllib.request import urlopen as uReq


import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
import datetime as dt
from matplotlib import rcParams

%matplotlib inline
```

```python
adp_2014 = pd.read_csv('adp_2014.csv')
```

```python
wr_adp_14 = adp_2014.loc[adp_2014['Position'] == 'WR']
te_adp_14 = adp_2014.loc[adp_2014['Position'] == 'TE']
qb_adp_14 = adp_2014.loc[adp_2014['Position'] == 'QB']
pk_adp_14 = adp_2014.loc[adp_2014['Position'] == 'PK']
rb_adp_14 = adp_2014.loc[adp_2014['Position'] == 'RB']
def_adp_14 = adp_2014.loc[adp_2014['Position'] == 'Def']
```

```python
rb_adp_14['PositionRank'] = rb_adp_14['Rank'].rank(ascending = True)
rb_adp_14.sort_values(by = 'PositionRank')
wr_adp_14['PositionRank'] = wr_adp_14['Rank'].rank(ascending = True)
wr_adp_14.sort_values(by = 'PositionRank')
te_adp_14['PositionRank'] = te_adp_14['Rank'].rank(ascending = True)
te_adp_14.sort_values(by = 'PositionRank')
qb_adp_14['PositionRank'] = qb_adp_14['Rank'].rank(ascending = True)
qb_adp_14.sort_values(by = 'PositionRank')
def_adp_14['PositionRank'] = def_adp_14['Rank'].rank(ascending = True)
def_adp_14.sort_values(by = 'PositionRank')
pk_adp_14['PositionRank'] = pk_adp_14['Rank'].rank(ascending = True)
pk_adp_14.sort_values(by = 'PositionRank')
```

```python
adp_14_ranked = pd.concat([wr_adp_14, rb_adp_14, qb_adp_14, pk_adp_14, te_adp_14, def_adp_14])
adp_14_ranked = adp_14_ranked.sort_values(by = 'Rank')
```

```python
rb_adp_14.to_csv('adp_rb_2014.csv', index = False)
wr_adp_14.to_csv('adp_wr_2014.csv', index = False)
te_adp_14.to_csv('adp_te_2014.csv', index = False)
qb_adp_14.to_csv('adp_qb_2014.csv', index = False)
def_adp_14.to_csv('adp_def_2014.csv', index = False)
pk_adp_14.to_csv('adp_pk_2014.csv', index = False)

adp_14_ranked.to_csv('adp_ranked_2014.csv', index = False)
```
