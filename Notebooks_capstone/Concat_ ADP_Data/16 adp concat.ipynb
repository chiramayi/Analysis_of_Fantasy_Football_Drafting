```python
import requests
from bs4 import BeautifulSoup
import pandas as pd
from urllib.request import urlopen as uReq


import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
import datetime as dt
from matplotlib import rcParams

%matplotlib inline
```

```python
adp_2016 = pd.read_csv('adp_2016.csv')
```

```python
wr_adp_16 = adp_2016.loc[adp_2016['Position'] == 'WR']
te_adp_16 = adp_2016.loc[adp_2016['Position'] == 'TE']
qb_adp_16 = adp_2016.loc[adp_2016['Position'] == 'QB']
pk_adp_16 = adp_2016.loc[adp_2016['Position'] == 'PK']
rb_adp_16 = adp_2016.loc[adp_2016['Position'] == 'RB']
def_adp_16 = adp_2016.loc[adp_2016['Position'] == 'Def']
```

```python
rb_adp_16['PositionRank'] = rb_adp_16['Rank'].rank(ascending = True)
rb_adp_16.sort_values(by = 'PositionRank')
wr_adp_16['PositionRank'] = wr_adp_16['Rank'].rank(ascending = True)
wr_adp_16.sort_values(by = 'PositionRank')
te_adp_16['PositionRank'] = te_adp_16['Rank'].rank(ascending = True)
te_adp_16.sort_values(by = 'PositionRank')
qb_adp_16['PositionRank'] = qb_adp_16['Rank'].rank(ascending = True)
qb_adp_16.sort_values(by = 'PositionRank')
def_adp_16['PositionRank'] = def_adp_16['Rank'].rank(ascending = True)
def_adp_16.sort_values(by = 'PositionRank')
pk_adp_16['PositionRank'] = pk_adp_16['Rank'].rank(ascending = True)
pk_adp_16.sort_values(by = 'PositionRank')
```

```python
adp_16_ranked = pd.concat([wr_adp_16, rb_adp_16, qb_adp_16, pk_adp_16, te_adp_16, def_adp_16])
adp_16_ranked = adp_16_ranked.sort_values(by = 'Rank')
```

```python
rb_adp_16.to_csv('adp_rb_2016.csv', index = False)
wr_adp_16.to_csv('adp_wr_2016.csv', index = False)
te_adp_16.to_csv('adp_te_2016.csv', index = False)
qb_adp_16.to_csv('adp_qb_2016.csv', index = False)
def_adp_16.to_csv('adp_def_2016.csv', index = False)
pk_adp_16.to_csv('adp_pk_2016.csv', index = False)

adp_16_ranked.to_csv('adp_ranked_2016.csv', index = False)
```
