```python
import requests
from bs4 import BeautifulSoup
import pandas as pd
from urllib.request import urlopen as uReq


import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
import datetime as dt
from matplotlib import rcParams

%matplotlib inline
```

```python
adp_2013 = pd.read_csv('adp_2013.csv')
```

```python
wr_adp_13 = adp_2013.loc[adp_2013['Position'] == 'WR']
te_adp_13 = adp_2013.loc[adp_2013['Position'] == 'TE']
qb_adp_13 = adp_2013.loc[adp_2013['Position'] == 'QB']
pk_adp_13 = adp_2013.loc[adp_2013['Position'] == 'PK']
rb_adp_13 = adp_2013.loc[adp_2013['Position'] == 'RB']
def_adp_13 = adp_2013.loc[adp_2013['Position'] == 'Def']
```

```python
rb_adp_13['PositionRank'] = rb_adp_13['Rank'].rank(ascending = True)
rb_adp_13.sort_values(by = 'PositionRank')
wr_adp_13['PositionRank'] = wr_adp_13['Rank'].rank(ascending = True)
wr_adp_13.sort_values(by = 'PositionRank')
te_adp_13['PositionRank'] = te_adp_13['Rank'].rank(ascending = True)
te_adp_13.sort_values(by = 'PositionRank')
qb_adp_13['PositionRank'] = qb_adp_13['Rank'].rank(ascending = True)
qb_adp_13.sort_values(by = 'PositionRank')
def_adp_13['PositionRank'] = def_adp_13['Rank'].rank(ascending = True)
def_adp_13.sort_values(by = 'PositionRank')
pk_adp_13['PositionRank'] = pk_adp_13['Rank'].rank(ascending = True)
pk_adp_13.sort_values(by = 'PositionRank')
```

```python
adp_13_ranked = pd.concat([wr_adp_13, rb_adp_13, qb_adp_13, pk_adp_13, te_adp_13, def_adp_13])
adp_13_ranked = adp_13_ranked.sort_values(by = 'Rank')
```

```python
rb_adp_13.to_csv('adp_rb_2013.csv', index = False)
wr_adp_13.to_csv('adp_wr_2013.csv', index = False)
te_adp_13.to_csv('adp_te_2013.csv', index = False)
qb_adp_13.to_csv('adp_qb_2013.csv', index = False)
def_adp_13.to_csv('adp_def_2013.csv', index = False)
pk_adp_13.to_csv('adp_pk_2013.csv', index = False)

adp_13_ranked.to_csv('adp_ranked_2013.csv', index = False)
```
