```python
import requests
from bs4 import BeautifulSoup
import pandas as pd
from urllib.request import urlopen as uReq


import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
import datetime as dt
from matplotlib import rcParams

%matplotlib inline
```

```python
adp_2015 = pd.read_csv('adp_2015.csv')
```

```python
wr_adp_15 = adp_2015.loc[adp_2015['Position'] == 'WR']
te_adp_15 = adp_2015.loc[adp_2015['Position'] == 'TE']
qb_adp_15 = adp_2015.loc[adp_2015['Position'] == 'QB']
pk_adp_15 = adp_2015.loc[adp_2015['Position'] == 'PK']
rb_adp_15 = adp_2015.loc[adp_2015['Position'] == 'RB']
def_adp_15 = adp_2015.loc[adp_2015['Position'] == 'Def']
```

```python
rb_adp_15['PositionRank'] = rb_adp_15['Rank'].rank(ascending = True)
rb_adp_15.sort_values(by = 'PositionRank')
wr_adp_15['PositionRank'] = wr_adp_15['Rank'].rank(ascending = True)
wr_adp_15.sort_values(by = 'PositionRank')
te_adp_15['PositionRank'] = te_adp_15['Rank'].rank(ascending = True)
te_adp_15.sort_values(by = 'PositionRank')
qb_adp_15['PositionRank'] = qb_adp_15['Rank'].rank(ascending = True)
qb_adp_15.sort_values(by = 'PositionRank')
def_adp_15['PositionRank'] = def_adp_15['Rank'].rank(ascending = True)
def_adp_15.sort_values(by = 'PositionRank')
pk_adp_15['PositionRank'] = pk_adp_15['Rank'].rank(ascending = True)
pk_adp_15.sort_values(by = 'PositionRank')
```

```python
adp_15_ranked = pd.concat([wr_adp_15, rb_adp_15, qb_adp_15, pk_adp_15, te_adp_15, def_adp_15])
adp_15_ranked = adp_15_ranked.sort_values(by = 'Rank')
```

```python
rb_adp_15.to_csv('adp_rb_2015.csv', index = False)
wr_adp_15.to_csv('adp_wr_2015.csv', index = False)
te_adp_15.to_csv('adp_te_2015.csv', index = False)
qb_adp_15.to_csv('adp_qb_2015.csv', index = False)
def_adp_15.to_csv('adp_def_2015.csv', index = False)
pk_adp_15.to_csv('adp_pk_2015.csv', index = False)

adp_15_ranked.to_csv('adp_ranked_2015.csv', index = False)
```
